```python
#Creating a copy of the categoricals dataset, just in case.
categoricals1 = categoricals.copy()
```

```python
#Dropping the columns that I see are not useful for the model.
categoricals1 = categoricals1.drop(['customer', 'response', 'education', 'employmentstatus', 'marital_status', 'sales_channel', 'renew_offer_type' ], axis=1)
```

```python
#Preparing OneHotEncoder
ohe=OneHotEncoder()
```

```python
#Running OneHotEncoder
ohe.fit_transform(categoricals1[['state', 'coverage', 'gender', 'location_code', 'policy_type', 'policy', 'vehicle_class', 'vehicle_size']]).toarray()
```

```python
#Saving the arrays in a variable
cat_array = ohe.fit_transform(categoricals1[['state', 'coverage','gender', 'location_code','policy_type', 'policy','vehicle_class', 'vehicle_size']]).toarray()
```

```python
#Checking the categories after OneHotEncoder
ohe.categories_
```

```python
#Saving them in a variable
cat_labels = ohe.categories_
```

```python
#Unifying the arrays in only one using hstack(), because the arrays are irregular
np.hstack(cat_labels)
```

```python
#Saving it in a variable
ctr_final = np.hstack(cat_labels)
```

```python
#Creating the dataframe with the encoded categoricals after OneHotEncoder
encoded_categoricals =pd.DataFrame(cat_array, columns = ctr_final)
```

```python
#Doing the X-y split while concatenating the numericals and encoded categoricals directly to X
X = pd.concat([numerical.drop(['total_claim_amount'], axis = 1), encoded_categoricals], axis=1)
y = numerical['total_claim_amount']
```

```python
#Preparing the train and test split before the scaling to prevent data leakage.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Calling the scaler.
scaler = StandardScaler()
```

```python
#Training the scaler
scaler.fit(X_train)
```

```python
#Saving the end results
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Calling the Linear Regression model.
lm = LinearRegression()
```

```python
#Training the algorithm.
lm.fit(X_train_scaled, y_train)
```

```python
#Getting and printing the results of the prediction

predict = lm.predict(X_test_scaled)
mse = mean_squared_error(y_test, predict, squared=True)
rmse = mean_squared_error(y_test, predict, squared=False)
mae = mean_absolute_error(y_test, predict)
print("R2_score:", round(r2_score(y_test, predict), 2))
print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
```

```python
#Plotting the predictions to check the results. As we can see, the dots start to scatter from the middle to the end of the chart,
#explaining the R2 score.

sns.regplot(x=y_test, y=predict, line_kws={"color": "red"}, scatter_kws={"color": "blue"})
```

```python
#Checking for outliers in the target.
sns.boxplot(numerical['total_claim_amount'])
```

```python
#Getting the data to further check the outliers

IQR = abs(np.quantile(numerical["total_claim_amount"], .25) - np.quantile(numerical["total_claim_amount"], .75)) * 1.5
lower_boundary = np.quantile(numerical["total_claim_amount"], .25) - IQR
upper_boundary = np.quantile(numerical["total_claim_amount"], .75) + IQR

# Outliers for this column are values smaller than lower_boundary or bigger than upper_boundary:
lower_boundary, upper_boundary
```

```python
#Printing the outliers
numerical[(numerical["total_claim_amount"] > 960)].sort_values("total_claim_amount")
```

```python
#Taking out the outliers from the numericals dataset to better up the model.
num_no_out = numerical[numerical['total_claim_amount'] < 960]
```

```python
#Concatenating the numerical and categorical dataframes.
df_final = pd.concat([num_no_out, encoded_categoricals], axis=1).reset_index(drop=True)
```

```python
#Dropping the NaNs
df_no_na = df_final.dropna()
```

```python
#Preparing the X-y split.
X = df_no_na.drop(['total_claim_amount'], axis = 1)
y = df_no_na['total_claim_amount']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Fitting the scaler again
scaler.fit(X_train)
```

```python
#Scaling and saving the results
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Fitting the model.
lm.fit(X_train_scaled, y_train)
```

```python
#Running the model again.
predict = lm.predict(X_test_scaled)
mse = mean_squared_error(y_test, predict, squared=True)
rmse = mean_squared_error(y_test, predict, squared=False)
mae = mean_absolute_error(y_test, predict)
print("R2_score:", round(r2_score(y_test, predict), 2))
print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
```

```python
#Plotting the model again to see the difference. AS we can see the model is now more clustered, and the R2 score is better.
sns.regplot(x=y_test, y=predict, line_kws={"color": "red"}, scatter_kws={"color": "blue"})
```

As we can see, the model got better by 0.02 points on RS2, and by 10000, 42 and 21 in the MSE, RMSE and MAE respectiveley by just removing the outliers from the dataframe. It has to be said, during the process we lost 459 rows, so the results may not be true to the reality.
